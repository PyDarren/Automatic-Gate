```python
import pandas as pd
import numpy as np
```

```python
path = 'C:/Users/pc/OneDrive/PLTTECH/Project/00_immune_age_project/'
formula_lvs = pd.read_csv(path+'Rawdata/formula_LVs.csv')
```

```python
formula_lvs
```

```python
coff_df = formula_lvs
```

```python
coff_df['LV1'] = coff_df['LV1'].apply(lambda x: -0.6259*x)
coff_df['LV2'] = coff_df['LV2'].apply(lambda x: 0.2941*x)
coff_df['LV3'] = coff_df['LV3'].apply(lambda x: -0.0356*x)
```

```python
coff_df['coefficient'] = coff_df['LV1'] + coff_df['LV2'] + coff_df['LV3']
```

```python
coff_df['coe_abs'] = coff_df['coefficient'].apply(lambda x : np.abs(x))
```

```python
coff_df.sort_values(by='coe_abs', ascending=False)
```
